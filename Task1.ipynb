```python
import matplotlib.pyplot as plt

plt.hist(
    np.random.rand(100000),
    density=True,
    bins=100,
    histtype="step",
    color="blue",
    label="rand"
)

plt.axis([0, 1, 0, 1.1])  # fits uniform rand data
plt.legend(loc="upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Create meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Equation Z = X^2 + Y^2
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and limits
ax.set_title("3D Surface: Z = X^2 + Y^2")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)

plt.show()
```

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv("Pokemon.csv", encoding="latin1")

# Select only the required columns
cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected = df[cols]

# Pearson correlation
pearson_corr = df_selected.corr(method='pearson')

# Spearman correlation
spearman_corr = df_selected.corr(method='spearman')

# Plot heatmaps
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[0])
axes[0].set_title("Pearson Correlation")

sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[1])
axes[1].set_title("Spearman Rank Correlation")

plt.tight_layout()
plt.show()
```
